## 2.1 - Calculando somatria de 1 a N

### <font color='gray'>Inicialmente iremos capturar o nmero do usurio pelo teclado</font>

```python
perguntaNumero = input("De qual nmero iremos calcular a somatria de 1 a N? ")
```

### <font color='gray'>- Vamos utilizar o comando "for" para calcular a soma dos nmeros
- Iremos utilizar tambm as funes: 
    - int() para converter de string para inteiro
    - range() para determinar o intervalor da repetio do For


</font>

```python
somatoria = 0
for i in range(1, int(perguntaNumero) + 1):
    somatoria = somatoria + i
```

### <font color='gray'> Apresentando o resultado final na tela:</font>
- Iremos utilizar tambm a funo: 
    - str() para converter de inteiro para string

```python
print("A somatria de 1 a " + perguntaNumero + "  " + str(somatoria))
```

## 3.2 - Escrevendo uma nova funo

### <font color='gray'> Funo que verifica se uma string  um palndromo </font>


```python
def verificarPalindromo(frase):
    # captura o tamanho da frase
    tamanho = len(frase)
    # executa a repetio para a metade do tamanho da string (//  a operao de diviso inteira)
    for i in range(tamanho//2):
        # vai percorrendo a string "para dentro" nos dois sentidos, de acordo com o valor de i
        if frase[i] != frase[(tamanho-1) - i]:
            #interrompe execuo caso encontre diferenas
            return False
        # se chegou ao final significa que a string  um palndromo
    return True
```

### <font color='gray'> Vamos testar com uma frase a ser digitada! </font>

```python
frase = input("digite string\n")
verificarPalindromo(frase)
```
